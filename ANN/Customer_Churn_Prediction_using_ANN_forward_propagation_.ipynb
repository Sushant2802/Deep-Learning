```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.drop(columns = ['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

```python
df['Geography'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['Geography', 'Gender'], drop_first=True)
```

```python
df.head()
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited'].values

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)
```

```python

import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(11, activation='sigmoid', input_dim = 11))   # add 11 node eith input dim 11
model.add(Dense(11, activation='sigmoid'))    # add 11 node
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python

model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python

history = model.fit(X_train_trf, y_train,
                    batch_size=50, epochs=100,
                    verbose=1, validation_split=0.2)
# side by side in the training data 0.2% data goes for validation_split
# also it gives side by side the accuracy of that validation set
```

```python

y_pred = model.predict(X_test_trf)
y_pred
```

```python

y_pred = y_pred.argmax(axis=-1)
y_pred
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
