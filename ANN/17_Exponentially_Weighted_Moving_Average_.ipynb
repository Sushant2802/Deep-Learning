```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DailyDelhiClimate.csv')[['date', 'meantemp']]
```

```python
df = df.head(100)
```

```python
df.head()
```

```python
plt.scatter(df['date'], df['meantemp'], color='red')
```

```python
x1 = df['meantemp'].ewm(alpha=0.5).mean()   # it means beta = 0.1
```

```python
x1
```

```python
df['ewma'] = x1
df.head()
```

```python
plt.scatter(df['date'], df['meantemp'], color='red')
plt.plot(df['date'], x1, color='black', label='ewma (alpha=0.9 & beta=0.1)', )
plt.legend()
plt.show()
```
